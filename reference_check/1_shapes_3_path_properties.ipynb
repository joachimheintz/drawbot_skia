# Shapes
## Path Properties

```python
from IPython.display import Image, SVG
from os import system
from drawbot_skia.drawbot import *

def drawShow(name):
    call = 'drawbot %s.py %s.png' % (name, name)
    system(call)
    return Image('%s.png' % name)
```

###   strokeWidth(value)
Sets stroke width.

```python
%%writefile strokeWidth.py
size(600,300)

fill(None)
stroke(0)
for i in range(20):
    strokeWidth(i)
    line((10, 10), (50, 250))
    translate(25, 0)
```

```python
drawShow('strokeWidth')
```

###   miterLimit(value) NOT WORKING?
Set a miter limit. Used on corner points.

```python
%%writefile miterLimit.py
size(400,400)

path = BezierPath()
path.moveTo((10, 10))
path.lineTo((100, 290))
path.lineTo((190, 10))
stroke(0)
fill(None)
strokeWidth(50)
drawPath(path)

translate(200, 0)
miterLimit(5)
drawPath(path)
```

```python
drawShow('miterLimit')
```

###   lineJoin(value)
Set a line join.
Possible values are miter, round and bevel.

```python
%%writefile lineJoin.py
size(600,300)

stroke(0)
fill(None)
strokeWidth(30)
miterLimit(30)

path = BezierPath()
path.moveTo((70, 10))
path.lineTo((100, 200))
path.lineTo((130, 10))
lineJoin("miter")
drawPath(path)

translate(200, 0)
lineJoin("round")
drawPath(path)

translate(200, 0)
lineJoin("bevel")
drawPath(path)
```

```python
drawShow('lineJoin')
```

###     lineCap(value)
Set a line cap.
Possible values are butt, square and round.

```python
%%writefile lineCap.py
size(400, 300)

stroke(0)
strokeWidth(50)
translate(50, 50)
lineCap("butt")
line((0, 0), (0, 200))

translate(100, 0)
lineCap("square")
line((0, 0), (0, 200))

translate(100, 0)
lineCap("round")
line((0, 0), (0, 200))
```

```python
drawShow('lineCap')
```

###   lineDash(*value) MIDDLE LINE IS WRONG!
Set a line dash with any given amount of lenghts. Uneven lenghts will have a visible stroke, even lenghts will be invisible.

```python
%%writefile lineDash.py
size(400,300)

stroke(0)
strokeWidth(50)
translate(50, 50)
lineDash(2, 2)
line((0, 0), (0, 200))

translate(100, 0)
lineDash(2, 10, 5, 5)
line((0, 0), (0, 200))

translate(100, 0)
lineDash(None)
line((0, 0), (0, 200))
```

```python
drawShow('lineDash')
```
