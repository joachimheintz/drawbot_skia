this is nothing but a transfer of the examples in drawbot_skia to jupyter notebooks

```python
from IPython.display import Image, SVG
from os import system
from drawbot_skia.drawbot import *

def drawShow(name):
    call = 'drawbot %s.py %s.png' % (name, name)
    system(call)
    return Image('%s.png' % name)
```

## arc

```python
%%writefile arc.py
size(200, 200)

path = BezierPath()
path.moveTo((100, 100))
path.arc((100, 100), 80, 230, 170, False)
path.arc((100, 100), 60, 30, 120, True)
path.closePath()

stroke(0)
fill(None)
strokeWidth(4)
drawPath(path)

# arc(center, radius, startAngle, endAngle, clockwise)
# Arc with center and a given radius, from startAngle to endAngle, 
# going clockwise if clockwise is True and counter clockwise if clockwise is False.

# arcTo(point1, point2, radius)
# Arc defined by a circle inscribed inside the angle specified by three points: 
# the current point, point1, and point2. The arc is drawn between the two points of the circle 
# that are tangent to the two legs of the angle.
```

```python
drawShow('arc')
```

## arcTo

```python
%%writefile arcTo.py
size(200, 200)

path = BezierPath()
path.moveTo((20, 20))
path.arcTo((20, 180), (180, 180), 30)
path.arcTo((180, 180), (110, 20), 30)
path.lineTo((110, 20))

stroke(0)
strokeWidth(8)
fill(None)

drawPath(path)
```

```python
drawShow('arcTo')
```

## blendMode

```python
%%writefile blendMode.py
modes = [
    'normal',
    'multiply',
    'screen',
    'overlay',
    'darken',
    'lighten',
    'colorDodge',
    'colorBurn',
    'softLight',
    'hardLight',
    'difference',
    'exclusion',
    'hue',
    'saturation',
    'color',
    'luminosity'
]

cellSize = 100
size(cellSize, cellSize * len(modes))
fill(0.5)
rect(0, 0, width(), height())

for mode in modes:
    blendMode(mode)

    fill(0, 0.4, 1)
    rect(0, 0, 0.75 * cellSize, 0.75 * cellSize)
    fill(1, 0, 0.4)
    rect(0.25 * cellSize, 0.25 * cellSize, 0.75 * cellSize, 0.75 * cellSize)
    translate(0, cellSize)
```

```python
drawShow('blendMode')
```

## clip

```python
%%writefile clip.py
size(200, 400)

path = BezierPath()
path.polygon((80, 28), (50, 146), (146, 152), (172, 78))

with savedState():
    clipPath(path)
    scale(200/512)
    image("../images/drawbot.png", (0, 0))

translate(0, 200)

with savedState():
    scale(200/512)
    image("../images/drawbot.png", (0, 0))
```

```python
drawShow('clip')
```
