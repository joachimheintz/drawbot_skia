## fontVariations

```python
%%writefile fontVariations.py
size(275, 400)

rect(0, 0, width(), height())
font("../fonts/MutatorSans.ttf")
fontSize(100)
fill(1)
text("ABCXYZ", (10, 15))
fontVariations(wght=600)
text("ABCXYZ", (10, 115))
variations = fontVariations(wdth=200)
assert {'wdth': 200, 'wght': 600.0} == variations, variations
text("ABCXYZ", (10, 215))
variations = fontVariations(wdth=1000, resetVariations=True)
assert {'wdth': 1000} == variations, variations
text("ABCXYZ", (10, 315))
```

```python
drawShow('fontVariations')
```

## language

```python
%%writefile language.py
size(600, 120)

proofText = ''
font('../fonts/SourceSerifPro-Regular.otf')
fontSize(34)

w, _ = textSize(proofText)
assert round(w, 1) == 498.0
text(proofText, (12, 74))

language("bg")
w, _ = textSize(proofText)
assert round(w, 1) == 537.7
text(proofText, (12, 33))
```

```python
drawShow('language')
```

## text shaping

```python
%%writefile textShaping.py
size(600, 500)

fill(1)
rect(0, 0, width(), height())
fill(0)
font("../fonts/IBMPlexSansArabic-Regular.otf")
fontSize(100)
openTypeFeatures(ss01=True, ss02=True, ss03=True)
openTypeFeatures(resetFeatures=True, ss03=True)
text("Aag0", (50, 50))
w, h = textSize("Aag0")
text("---", (50 + w, 50))

openTypeFeatures(resetFeatures=True)
fontSize(60)
s = " abcde "
text(s, (width() - 50, 150))

text("Alfabet", (width() / 2, 220))
text("", (width() / 2, 290))

font("Helvetica")  # simply test loading an installed font
```

```python
drawShow('textShaping')
```
